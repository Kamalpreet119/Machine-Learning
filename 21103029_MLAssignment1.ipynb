```python
pandasSeries=pd.Series(Numpy_Series).unique() #To get all Unique Values
print(pandasSeries)
```

```python
def LabelEncoder(NumpyArray):
    for i in range(len(NumpyArray)):
        if(NumpyArray[i]=="BENIGN"):
            NumpyArray[i]=1
        elif(NumpyArray[i]=="DoS"):
            NumpyArray[i]=2
        elif(NumpyArray[i]=="PortScan"):
            NumpyArray[i]=3
        elif(NumpyArray[i]=="Bot"):
            NumpyArray[i]=4
        elif(NumpyArray[i]=="Infiltration"):
            NumpyArray[i]=5
        elif(NumpyArray[i]=="WebAttack"):
            NumpyArray[i]=6
        elif(NumpyArray[i]=="BruteForce"):
            NumpyArray[i]=7
        else:
            NumpyArray[i]=-1
    return NumpyArray
```

```python
arr=LabelEncoder(Numpy_Series)
print(arr)
```

```python
def LabelEncoder1(arr):
    from sklearn.preprocessing import LabelEncoder
    Le=LabelEncoder()
    arr=Le.fit_transform(arr)
    return arr
```

### Simple Imputer

```python
from sklearn.impute import SimpleImputer
imputer1=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer2=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer3=SimpleImputer(missing_values=np.nan,strategy='constant',fill_value=0)
imputer4=SimpleImputer(missing_values=np.nan,strategy='median')
```

```python
imputer1.fit(df1)
imputer2.fit(df1)
imputer3.fit(df1)
imputer4.fit(df1)
df10=imputer1.transform(df1)
df11=imputer2.transform(df1)
df12=imputer3.transform(df1)
df13=imputer4.transform(df1)
```
